```python
#Doubly linked list
#creating node which contain data and store the address of another variable
class node:
  def __init__(self,data):
    self.data= data
    self.next = None
    self.prev = None

#creating class of doubly linked list
class Dll:
  def __init__(self):
    self.head = None
  
#in DOUBLE linked list we can traversal from the beginning and traversal at the end
  def forward_traversal(self):
    if self.head is None:
      print("dll is empty")
    else:
      a = self.head
      while a is not None:
        print(a.data,end=" ")
        a = a.next
    print()
  
  #traversal through end
  def backward_traversal(self):
    a = self.head
    while a.next is not None:
      a = a.next
    while a is not None:
      print(a.data,end=" ")
      a = a.prev
    print()

#dll can insert elements at the beginning,end and specific postition
  def inserting_elements_at_beginning(self,data):
    nb = node(data)
    a = self.head
    nb.next = a
    a.prev = nb
    self.head = nb
  
  def inserting_element_at_end(self,data):
    ne = node(data)
    a = self.head
    while a.next is not None:
        a =a.next
    ne.prev= a.next
    a.next = ne
  
  def inserting_element_at_specific_position(self,data,position):
    ns = node(data)
    a = self.head
    for i in range(1,position-1):
      a = a.next
    ns.prev = a
    ns.next = a.next
    a.next.prev = ns
    a.next = ns
  
  def deletion_at_beginning(self):
    a = self.head
    self.head = a.next
    a.next = None
    self.head.prev= None
  
  def deletion_at_end(self):
    a = self.head.next
    prev = self.head
    while a.next is not None:
      a = a.next
      prev= prev.next
    prev.next = None
    a.prev = None
  
  def deletion_at_specific_position(self,position):
    a = self.head.next
    before = self.head
    for i in range(1,position-1):
      a = a.next
      before = before.next
    before.next = a.next
    a.next.prev = before
    a.next = None
    a.prev = None

dll = Dll()
n1 = node(20)
dll.head = n1
n2 = node(30)
n1.next = n2
n2.prev =n1
n3 = node(40)
n2.next = n3
n3.prev = n2
n4 = node(50)
n3.next =n4
n4.prev = n3
n5 = node(60)
n4.next = n5
n5.prev =n4 

#traversal from beginning
dll.forward_traversal()
#traversal through end
dll.backward_traversal()

#adding the elements at the beginning
dll.inserting_elements_at_beginning(5)
dll.forward_traversal()

#adding the elements at the end
dll.inserting_element_at_end(70)
dll.forward_traversal()

#inserting elements at specific position
dll.inserting_element_at_specific_position(100,2)
dll.forward_traversal()

print("deletion elements at beginnning")
dll.deletion_at_beginning()
dll.forward_traversal()

dll.deletion_at_end()
dll.forward_traversal()

dll.deletion_at_specific_position(2)
dll.forward_traversal()
```
